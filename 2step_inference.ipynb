```python
# In this notebook we display the code used to estimate the 2-step GLM parameters.
# Importing the necessary libraries 
import sys
#Importing the precreated helper tools for fitting the data to a linear-nonlinear model 
sys.path.append('./helpers/')

from data_manip import *
from temporal_bases import *
from compiled_models import *

#tensorflow needed to run the convolutional 2D model and the loss function of the model
import tensorflow as tf
# spike and stimulus stored in numpy arrays, raw data must be unpickled 
import numpy as np
import pickle5 as pkl
import os
#matplotlib needed for visualizing the loss function
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Testing data must have the following format
# stimulus: (nx, ny, nt_test, 1) where
#nx and ny are the spatial dimensions and nt_test is the number of time frames (each time frame being 1,67ms)
# spikes: (n_cells, nt_test, n_rep_test) where 
# n_cells is the number of neurons, nt_test is the number of time frames (each time frame being 1,67ms), and 
#n_rep_test is the number of stimulus repeats.
# Unpickling and loading the training data
with open('data/' + 'data_bars_OFF_repeated' + '.pkl', 'rb') as f:
    data_train = pkl.load(f)
    
#lets view the data
print(data_train)     
```

```python
#data is a dictionary consisting of 2 keys: spikes and stimulus. We only need the spikes here as the 
#2step inference is for modelling firing rates, which is essentially the spike rate. 
spikes_test = data_train['spikes']
#To reduce required run time, we can limit the number of time frames to to 2000.

spikes = data_train['spikes'][:,2000:,:]
print(f'Training data:\n',
      f'spikes shape: {spikes.shape}')
# just to plot the spikes. We randomly are looking at a visualization of the second cell for all repeats in the first
# 5 time frames

for i in range(0,5):
    plt.plot(spikes[1][i])
    plt.xlabel('repeats')
    plt.yticks([])
    plt.show()
```

```python
# Assigning the dimensions of the spikes to the respective variables 
n_cells, nt, n_rep = spikes.shape

# No pretrained model will be used since a novel 2 step inference model is being created. The coefficients 
# will be stored in a separate folder within the model_weights directory called Int. 
load_pretrained = False
path_weights = './model_weights/Int/'

# In order for the model to run, certain predetermined values must be added that act as constants for 
# effective parameter estimation. These are temporal constants, which allow for better modelling of 
# the neuron firing rates due to its own (self activity) as well as the influence of neighbouring cell activities
# (coupling activity) over time. 
first_peak_coupl, last_peak_coupl, streach_coupl, n_basis_coupl, nt_integ_coupl = 0, 15, 5, 4, 25
last_peak_self, streach_self, n_basis_self, nt_integ_self = 20, 1, 5, 25

# The coupling parameters act as transformations to a raised cosine function to model 
# the influence of the activity of other cells on the investigated cell spike activity.
coupl_basis = raised_cosine_basis(first_peak_coupl, last_peak_coupl, streach_coupl, n_basis_coupl, nt_integ_coupl)

# Training parameters
nepochs = 200
batch_size = 1024

# The model will also account for the loss function. 
losses = np.zeros((n_cells, nepochs))
models = []

for cell in range(n_cells):
    
    # Combining all the parameters, we are able to create the model 
    n_basis_coupl, nt_integ_coupl = coupl_basis.shape
    n_basis_self, nt_integ_self = self_basis.shape
    nt_integ = np.max([nt_integ_coupl, nt_integ_self])
    nt_train = nt - nt_integ
    # Using the parameters, we are able to predict the constant that models the spike activity of a neuron 
    # without influence from activity of other neurons (self).
    self_basis, tau_r = self_basis_gen(last_peak_self, streach_self, n_basis_self, nt_integ_self, cell, spikes)

    # Combining the coupling and self constants for each cell, we can build a dataset.
    dataset = build_interaction_dataset(spikes, cell, coupl_basis, self_basis, tau_r)
    dataset = tf.data.Dataset.from_tensor_slices(dataset).shuffle(nt_train)
    dataset = dataset.batch(batch_size)

    # We can then input the training dataset and the parameters into a compiled interaction model

    temp_model = compiled_interaction_model(nt_train=nt_train, n_cells=n_cells, n_basis_coupl=n_basis_coupl, 
                                                   n_basis_self=n_basis_self, l2_stim_field=1e-7, l2_lapl_reg=0,
                                                   l1_coupl=1e-5, l2_coupl=0, l1_self=1e-5, l2_self=0)
    
    # Loading in pretrained model weights for calculating psth values
    if load_pretrained == True:
        temp_model.load_weights(path_weights+'cell'+str(cell))
    else:
        log_psth = np.array(np.log(np.mean(spikes[cell,nt_integ:,:,np.newaxis], 1)+1e-12), dtype='float32')
        init_weights = [log_psth-np.mean(log_psth), np.mean(log_psth)[np.newaxis]]
        temp_model.get_layer('stim_field').set_weights(init_weights)
        
    #Obtaining and storing loss functions for each cell
    loss_epoch = []
    for epoch in range(nepochs):
        loss_value = 0
        for stim_train, coupl_train, self_train, refr_train, spikes_train in dataset:
            loss_value += temp_model.train_on_batch([stim_train, coupl_train, self_train, refr_train], spikes_train, sample_weight=None, reset_metrics=True)
        if (epoch+1)%10==0: print(f'Epoch: {epoch+1}, loss: {loss_value/(nt_train*n_rep)}')
        loss_epoch += [loss_value/(nt_train*n_rep)]
   #plotting the loss function for each cell     
    plt.plot(loss_epoch)
    plt.show()
    losses[cell,:] = loss_epoch

    # Store model
    models += [temp_model]
```

```python
# Finally, we save the predicted constants in a separate folder for each cell
path_weights = './model_weights/Int/'

if not os.path.exists(path_weights):
    os.makedirs(path_weights)

for cell in range(n_cells):
    model = models[cell]
    model.save_weights(path_weights+'cell'+str(cell))
```

```python
# We repeat the process to extract the filters from the model, and the weights and biases for the coupling constants
# As well as the self constants. 

path_filters = './model_filters/Int/'

stimulus_fields, bias_list, coupl_filters, self_filters, tau_r_list = [], [], [], [], []
for cell in range(n_cells):
    # Weights and biases
    stim_field, bias = models[cell].get_layer(name='stim_field').weights
    coupling_coeffs = models[cell].get_layer(name='coupling_filters').weights[0]
    self_coeffs = models[cell].get_layer(name='self_filter').weights[0]
    
    # Stimulus model
    stimulus_fields += [stim_field.numpy()]
    bias_list += [bias.numpy()]
    
    # Couplings
    coupl_filters += [np.matmul(np.reshape(coupling_coeffs.numpy(), [n_cells-1, n_basis_coupl]), coupl_basis)]
    
    # Self coupling and refractory period
    self_basis, tau_r = self_basis_gen(last_peak_self, streach_self, n_basis_self,
                                   nt_integ_self, cell, spikes)
    self_filters += [np.matmul(self_coeffs.numpy().transpose(), self_basis)]
    tau_r_list += [tau_r]
    
if not os.path.exists(path_filters):
    os.makedirs(path_filters)

with open(path_filters+'filters.pkl', 'wb') as f:
    pkl.dump((bias_list, stimulus_fields, coupl_filters, self_filters, tau_r_list), f)
```

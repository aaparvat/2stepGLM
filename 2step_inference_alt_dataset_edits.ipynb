```python
# Some parameters
n_cells, nt, n_rep = spikes.shape

# Pretrained model
load_pretrained = False
path_weights = './model_weights/Int/'

# Temporal bases parameters
first_peak_coupl, last_peak_coupl, streach_coupl, n_basis_coupl, nt_integ_coupl = 0, 15, 5, 4, 25
last_peak_self, streach_self, n_basis_self, nt_integ_self = 20, 1, 5, 20

# Coupling basis
coupl_basis = raised_cosine_basis(first_peak_coupl, last_peak_coupl, streach_coupl, n_basis_coupl, nt_integ_coupl)

# Training parameters
nepochs = 200
batch_size = 1024

# Storage
losses = np.zeros((n_cells, nepochs))
models = []

for cell in range(n_cells):
    # Self coupling basis
    self_basis, tau_r = self_basis_gen(last_peak_self, streach_self, n_basis_self, nt_integ_self, cell, spikes)

    # Dataset
    dataset = build_interaction_dataset(spikes, cell, coupl_basis, self_basis, tau_r)
    dataset = tf.data.Dataset.from_tensor_slices(dataset).shuffle(nt_train)
    dataset = dataset.batch(batch_size)

    # Create the model
    n_basis_coupl, nt_integ_coupl = coupl_basis.shape
    n_basis_self, nt_integ_self = self_basis.shape
    nt_integ = np.max([nt_integ_coupl, nt_integ_self])
    nt_train = nt - nt_integ

    temp_model = compiled_interaction_model(nt_train=nt_train, n_cells=n_cells, n_basis_coupl=n_basis_coupl, 
                                                   n_basis_self=n_basis_self, l2_stim_field=1e-7, l2_lapl_reg=0,
                                                   l1_coupl=1e-5, l2_coupl=0, l1_self=1e-5, l2_self=0)
    
    # Load pretrain model weights
    if load_pretrained == True:
        temp_model.load_weights(path_weights+'cell'+str(cell))
    else:
        log_psth = np.array(np.log(np.mean(spikes[cell,nt_integ:,:,np.newaxis], 1)+1e-12), dtype='float32')
        init_weights = [log_psth-np.mean(log_psth), np.mean(log_psth)[np.newaxis]]
        temp_model.get_layer('stim_field').set_weights(init_weights)
    
    loss_epoch = []
    for epoch in range(nepochs):
        loss_value = 0
        for stim_train, coupl_train, self_train, refr_train, spikes_train in dataset:
            loss_value += temp_model.train_on_batch([stim_train, coupl_train, self_train, refr_train], spikes_train, sample_weight=None, reset_metrics=True)
        # Print and store loss
        if (epoch+1)%10==0: print(f'Epoch: {epoch+1}, loss: {loss_value/(nt_train*n_rep)}')
        loss_epoch += [loss_value/(nt_train*n_rep)]
        
    plt.plot(loss_epoch)
    plt.show()
    losses[cell,:] = loss_epoch

    # Store model
    models += [temp_model]
```

```python
# Save the weights of the models
path_weights = './model_weights/Int/'

if not os.path.exists(path_weights):
    os.makedirs(path_weights)

for cell in range(n_cells):
    model = models[cell]
    model.save_weights(path_weights+'cell'+str(cell))
```

```python
# Extract filters from the model, and the weights and biases

path_filters = './model_filters/Int/'

stimulus_fields, bias_list, coupl_filters, self_filters, tau_r_list = [], [], [], [], []
for cell in range(n_cells):
    # Weights and biases
    stim_field, bias = models[cell].get_layer(name='stim_field').weights
    coupling_coeffs = models[cell].get_layer(name='coupling_filters').weights[0]
    self_coeffs = models[cell].get_layer(name='self_filter').weights[0]
    
    # Stimulus model
    stimulus_fields += [stim_field.numpy()]
    bias_list += [bias.numpy()]
    
    # Couplings
    coupl_filters += [np.matmul(np.reshape(coupling_coeffs.numpy(), [n_cells-1, n_basis_coupl]), coupl_basis)]
    
    # Self coupling and refractory period
    self_basis, tau_r = self_basis_gen(last_peak_self, streach_self, n_basis_self,
                                   nt_integ_self, cell, spikes)
    self_filters += [np.matmul(self_coeffs.numpy().transpose(), self_basis)]
    tau_r_list += [tau_r]
    
if not os.path.exists(path_filters):
    os.makedirs(path_filters)

with open(path_filters+'filters.pkl', 'wb') as f:
    pkl.dump((bias_list, stimulus_fields, coupl_filters, self_filters, tau_r_list), f)
```
